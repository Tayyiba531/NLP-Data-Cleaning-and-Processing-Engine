It reduce the words to their base form. e.g words like connecting or connected. It lower the number of unique words in our dataset; therefore reducing the size and complexity of our data. It is important step for preparing our data properly for machine learning.

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connection', 'connects']
```

```python
for t in connect_tokens:
    print(t, ":", ps.stem(t))
```

```python
learn_tokens = ['learnt', 'learned', 'learning', 'learns', 'learner',]
```

```python
for t in learn_tokens:
    print(t, ":", ps.stem(t))
```
