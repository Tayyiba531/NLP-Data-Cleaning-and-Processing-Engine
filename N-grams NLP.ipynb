We can break our text down into n-grams to check our preprocessing, explore the contents of our data, or creat new feaures for machine learning. An n-gram is simply a sequence of neighbouring n words (or tokens), where n can be any number.

```python
import nltk
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significant', 
           'advancements', 'in', 'natural', 'language', 'processing', 'computer', 'vision', 'and', 'other', 
           'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated', 
           'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought', 
           'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of', 'deep', 
           'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful', 'capable', 'of', 
           'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from', 'it', 'in', 'ways', 
           'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai', 'is', 'increasingly', 
           'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from', 'healthcare', 'to', 
           'finance', 'to', 'transportation', 'and', 'and', 'its', 'impact', 'is', 'only', 'set', 'to', 
           'grow', 'in', 'the', 'years', 'to', 'come']

print(tokens)
```

```python
#Uni-Gram: n=1
unigrams = (pd.Series(nltk.ngrams(tokens,1)).value_counts())
print(unigrams)
```

```python
unigrams[0:10].sort_values().plot.barh(color='lightsalmon',width=0.9, figsize=(12,8))
plt.title("10 Most frequently Values")
plt.xlabel("#No")
plt.ylabel("Unigrams")
```

```python
#bi-Gram: n=2
bigrams = (pd.Series(nltk.ngrams(tokens,2)).value_counts())
print(bigrams)
```

```python
bigrams[0:10].sort_values().plot.barh(color='blue',width=0.9, figsize=(12,8))
plt.title("10 Most frequently Values")
plt.xlabel("#No")
plt.ylabel("bigrams")
```

```python
#Tri-Gram: n=3
trigrams = (pd.Series(nltk.ngrams(tokens,3)).value_counts())
print(trigrams)
```

```python
bigrams[0:10].sort_values().plot.barh(color='red',width=0.9, figsize=(12,8))
plt.title("10 Most frequently Values")
plt.xlabel("#No")
plt.ylabel("trigrams")
```
